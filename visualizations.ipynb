```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
serieA = pd.read_csv('data/serieA/results.csv')

roma = serieA[serieA['Team'] == 'Roma']

roma
```

```python
# Create readable row labels and format nicely
roma_viz = roma.copy()
roma_viz['Match'] = roma_viz['Opponent'] + ' (' + roma_viz['Venue'] + ')'
roma_viz = roma_viz.set_index('Match')

plt.figure(figsize=(8, 6))
sns.heatmap(roma_viz[['Points_cur', 'Points_prev', 'Differential', 'Cumulative']], 
            annot=True,      # Show numbers
            cmap='RdYlGn',   # Red-Yellow-Green
            center=0,        # Center at 0
            fmt='.1f',       # 1 decimal place
            linewidths=1)
plt.title('Roma Performance Heatmap')
plt.tight_layout()
plt.show()
```
